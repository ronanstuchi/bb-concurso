```python
# convertendo coluna "Price" para double.
# https://stackoverflow.com/questions/8421922/how-do-i-convert-a-currency-string-to-a-floating-point-number-in-python
car_sales["Price"] = car_sales["Price"].replace('[\$\,]', '', regex=True).astype(float)
car_sales
```
